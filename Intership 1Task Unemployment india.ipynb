```python
upper = per75th +(1.5*IQR_unem)

lower = per25th -(1.5*IQR_unem)

print(upper,lower)
```

```python
outliers=df_unemp[(df_unemp["Estimated Unemployment Rate (%)"]> upper) | (df_unemp["Estimated Unemployment Rate (%)"]<lower) ]
outliers.shape
```

## Visualize Data

```python
avg_unemployment_rate=df_unemp.groupby('Region')['Estimated Unemployment Rate (%)'].mean()
```

```python
avg_unemployment_rate
```

### Average Unemployment Rate by Region

```python
plt.figure(figsize=(12,7))
sns.barplot(x=avg_unemployment_rate.index ,y=avg_unemployment_rate.values,color='#0095c7')
plt.xlabel('Region')
plt.ylabel('Unemployment percentage')
plt.title('Average Unemployment Rate by Region')
plt.xticks(rotation=90)
plt.tight_layout()
```

### Relationship Between Unemployment Rate and Labour Participation Rate and Estimated Employed

```python
sns.pairplot(df_unemp,height=4)
plt.show()
```

```python
total_nums=df_unemp.groupby("Date")["Estimated Unemployment Rate (%)"].sum()
total_nums
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x="Date", y="Estimated Unemployment Rate (%)", data=df_unemp,hue="Area")
plt.title("Total Unemployment Rate Over Time ")
plt.xlabel("Date")
plt.ylabel("Unemployment Percentage (%)")
plt.show()
```

### Unemployment Rate by region


```python
plt.figure(figsize=(12,7))
sns.boxplot(x=df_unemp["Area"] ,y=df_unemp["Estimated Unemployment Rate (%)"])
plt.xlabel('Region')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment Rate by Region')
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(x=df_unemp["Area"] ,y=df_unemp["Estimated Labour Participation Rate (%)"])
plt.xlabel('Region')
plt.ylabel('Labour Participation Rate')
plt.title('Labour Participation Rate by Region')
plt.xticks(rotation=90)
plt.tight_layout()
```
